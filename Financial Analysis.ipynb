### Data Visualization

```python
sns.countplot(x = 'Segment', data = df, linewidth = 3, palette = "Set2", edgecolor = 'black')
plt.show;
```

```python
sns.countplot(x = 'Manufacturing Price', data = df, palette = "mako", linewidth = 2, edgecolor = "black")
plt.show();
```

```python
# Analyzing the products and discount band by segment.

plt.figure(figsize = (18,6))
plt.subplot(1,2,1)
sns.countplot(x = df['Segment'], hue = df['Product'], palette = 'rocket', linewidth = 3, edgecolor = "white")
plt.title("Products")

plt.subplot(1,2,2)
sns.countplot(x = df['Segment'], hue = df['Discount Band'], palette = 'hot', linewidth = 3, edgecolor = 'white')
plt.title("Discount Band")
plt.show;
```

```python
# Counting the years the products were sold

sns.countplot(x = df['Year'], palette = 'icefire', linewidth = 2, edgecolor = "black")
plt.show()
```

```python
# Analyizng the products by date sold

plt.figure(figsize=(10, 6))
sns.histplot(df['Date'], color = 'darkred')
plt.title('Products By Date Sold')
plt.show()
```

```python
# Analyzing the profit obtained in each month

sns.barplot(data = df, x = 'Profit', y = 'Month Name', color = 'tomato')
plt.title('Months Products were Sold By Profit')
plt.xlabel('Profit')
plt.ylabel('Month Name');
```

```python
# Analyzing the profit made by each product.

plt.scatter(data = df, x = 'Product', y = 'Profit', color = 'brown')
plt.title('Profit By Products')
plt.xlabel('Product')
plt.ylabel('Profit');
```

```python
# Analzing the profit made by each segment.

plt.scatter(data = df, x = 'Segment', y = 'Profit', color = 'brown')
plt.title('Profit By Segment')
plt.xlabel('Segment')
plt.ylabel('Profit');
```

```python
# Analyzing the products by sale price.

plt.figure(figsize = (16,8))
sns.pointplot(x = "Product", y = "Sale Price", data = df, linestyles = "--", capsize = .3, color = 'orangered')
plt.title("Sale Price By Product")
plt.show()
```

```python
# Analyzing segment by year and discount band by products.

plt.figure(figsize = (18,6))
plt.subplot(1,2,1)
sns.countplot(hue = df['Segment'], x = df['Year'], palette = 'viridis', linewidth = 2, edgecolor = "black")
plt.title("DURATION")

plt.subplot(1,2,2)
sns.countplot(hue = df['Product'], x = df['Discount Band'], palette = 'seismic', linewidth = 2, edgecolor = "black")
plt.title("DISCOUNT BAND BY PRODUCT")
plt.show();
```
